# Variant 8

\begin{align*}
\begin{cases}
13.1x_1 + 9.2x_2 &= 173 \\
8.3x_1 + 21x_2 &= 271 \\
3.2x_1 - 8.3x_2 &= 34 \\
\end{cases}
\end{align*}

## Task 1

```python
import numpy as np
import matplotlib.pyplot as plt

x_1 = np.linspace(0, 10, 100)
equations_to_plot = [
    (173 - 13.1 * x_1) / 9.2,
    (271 - 8.3 * x_1) / 21,
    (3.2 * x_1 - 34) / 8.3
]

equations_notations = [
    '$13.1x_1 + 9.2x_2 = 173$',
    '$8.3x_1 + 21x_2 = 271$',
    '$3.2x_1 - 8.3x_2 = 34$'
]

equations_count = len(equations_to_plot)

fig, axes = plt.subplots(ncols = equations_count, figsize=(10, 5))

fig.suptitle('Task 1')

for i in range(equations_count):
    axes[i].set(xlim=(0, 10), ylim=(-10, 30), title=equations_notations[i])

axes[1].set_yticklabels([])
axes[2].set_yticklabels([])

axes[0].plot(x_1, equations_to_plot[0], linestyle='--', color='red')
axes[1].plot(x_1, equations_to_plot[1], linestyle='dotted', color='blue')
axes[2].plot(x_1, equations_to_plot[2], linestyle='-.', color='purple')

plt.plot()
```
