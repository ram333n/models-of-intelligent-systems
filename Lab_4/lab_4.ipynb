# Variant 8

\begin{align*}
\begin{cases}
13.1x_1 + 9.2x_2 &= 173 \\
8.3x_1 + 21x_2 &= 271 \\
3.2x_1 - 8.3x_2 &= 34 \\
\end{cases}
\end{align*}

## Task 1

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-whitegrid')

x_1 = np.linspace(0, 10, 100)
equations_to_plot = [
    (173 - 13.1 * x_1) / 9.2,
    (271 - 8.3 * x_1) / 21,
    (3.2 * x_1 - 34) / 8.3
]

equations_notations = [
    '$13.1x_1 + 9.2x_2 = 173$',
    '$8.3x_1 + 21x_2 = 271$',
    '$3.2x_1 - 8.3x_2 = 34$'
]

equations_count = len(equations_to_plot)

fig, axes = plt.subplots(ncols = equations_count, figsize=(10, 5))

fig.suptitle('Task 1')

for i in range(equations_count):
    axes[i].set(xlim=(0, 10), ylim=(-10, 30), title=equations_notations[i])

axes[1].set_yticklabels([])
axes[2].set_yticklabels([])

axes[0].plot(x_1, equations_to_plot[0], linestyle='--', color='red')
axes[1].plot(x_1, equations_to_plot[1], linestyle='dotted', color='blue')
axes[2].plot(x_1, equations_to_plot[2], linestyle='-.', color='purple')

plt.plot()

plt.savefig('task_1.png')
plt.savefig('task_1.jpg')
plt.savefig('task_1.svg')
```

## Task 2

```python
equations_rhs_parts = [
    np.array([
        [13.1, 9.2],
        [8.3, 21],
    ]),
    np.array([
        [13.1, 9.2],
        [3.2, -8.3],
    ]),
    np.array([
        [8.3, 21],
        [3.2, -8.3],
    ])
]

equations_lhs_parts = [
    np.array([173, 271]),
    np.array([173, 34]),
    np.array([271, 34]),
]

intersections = []

for i in range(len(equations_rhs_parts)):
    intersections.append(np.linalg.solve(equations_rhs_parts[i], equations_lhs_parts[i]))

x_1_plot = np.linspace(0, 30, 100)
equations_to_plot = [
    (173 - 13.1 * x_1_plot) / 9.2,
    (271 - 8.3 * x_1_plot) / 21,
    (3.2 * x_1_plot - 34) / 8.3
]

fig = plt.figure(figsize=(10, 10), dpi=100)
axes = plt.axes()

axes.set_xlim((0, 30))
axes.set_xlabel('$x_1$')
axes.set_ylim((-10, 20))
axes.set_ylabel('$x_2$')
axes.set_title('Task 2(All lines in one figure)')

axes.plot(x_1_plot, equations_to_plot[0], linestyle='--', color='red', label='f[0]')
axes.plot(x_1_plot, equations_to_plot[1], linestyle='dotted', color='blue', label='f[1]')
axes.plot(x_1_plot, equations_to_plot[2], linestyle='-.', color='purple', label='f[2]')

axes.plot(*intersections[0], marker='o', color='r')
axes.plot(*intersections[1], marker='o', color='g')
axes.plot(*intersections[2], marker='o', color='y')

line_annotations_coords = [
    (1, 18),
    (25, 3.5),
    (3, -2.5)
]
rotations = [-55, -21, 23]

for i in range(equations_count):
    axes.text(
        *line_annotations_coords[i],
        equations_notations[i],
        rotation=rotations[i],
        rotation_mode='anchor',
        transform_rotates_text=True
    )

axes.annotate('The intersection point of\n f[0] and f[1]', ha='center', xy=intersections[0], xytext=(9, 12), arrowprops=dict(arrowstyle="->"))
axes.annotate('The intersection point of\n f[0] and f[2]', ha='center', xy=intersections[1], xytext=(10, -5), arrowprops=dict(arrowstyle="->"))
axes.annotate('The intersection point of\n f[1] and f[2]', ha='center', xy=intersections[2], xytext=(20, -3), arrowprops=dict(arrowstyle="->"))

y = np.maximum(equations_to_plot[0], equations_to_plot[2])
y_1 = np.maximum(equations_to_plot[1], y)
plt.fill_between(x_1_plot, y, y_1, color='lightgrey', alpha=0.5)

axes.legend()

plt.savefig('task_2.png')
plt.savefig('task_2.jpg')
plt.savefig('task_2.svg')
```
